```python
empty_list = []
index = 0
x_lat = []
y_lon = []
for (x,y) in lat_long_list:
    try:
        location = reverse((x, y), language='en', exactly_one=True)
        value = list(location.raw.items())
        empty_list.append(value)
        x_lat.append(x)
        y_lon.append(y)
        index += 1
        print(index)
    except:
        print('error'+str(index))
        pass
        
        
```
