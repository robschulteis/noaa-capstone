```python
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup as BS
from IPython.core.display import HTML
from datetime import datetime, timedelta
```

```python
week_ending_str = '2022-06-05'
week_ending = datetime.strptime(week_ending_str, '%Y-%m-%d')
modifier = timedelta(7)

show = []
weekblah = []
diff = []
avg = []
seats = []
perfs = []
cap = []
capdiff = []
week = []
i=1
week = []
while i in range(1,3):
    print(i)
    print(week_ending)
    #week.append(week_ending)                                                         #week
    
    url = f'https://www.playbill.com/grosses?week={str(week_ending.date())}'
    print(url)
    #response = requests.get(url)
    data = requests.get(url).text
    soup = BS(data, 'html.parser')
    table = soup.find('table', class_='bsp-table')
    #df = pd.DataFrame(columns=['Show', 'This Week Gross', 'Diff $', 'Avg Ticket', 'Seats Sold', 'Perfs', '% Cap', 'Diff % cap'])

    for row in table.tbody.find_all('tr'):    
        # Find all data for each column
        columns = row.find_all('td')
        if(columns != []):
            Show = columns[0].text.strip()
            show.append(Show)
            Week = columns[1].text.strip()
            weekblah.append(Week)
            Diff = columns[2].text.strip()
            diff.append(Diff)
            AvgTicket = columns[3].text.strip()
            avg.append(AvgTicket)
            Seats = columns[4].text.strip()
            seats.append(Seats)
            Perfs = columns[5].text.strip()
            perfs.append(Perfs)
            CapPerc = columns[6].text.strip()
            cap.append(CapPerc)
            CapPercDiff = columns[7].text.strip()
            capdiff.append(CapPercDiff)
            week.append(week_ending)

        #df = df.append({'Show': Show,  'This Week Gross': Week, 'Diff $': Diff, 'Avg Ticket': AvgTicket, 'Seats Sold': Seats, 'Perfs': Perfs, '% Cap': CapPerc, 'Diff % cap': CapPercDiff}, ignore_index=True)

    i +=1
    week_ending = week_ending - modifier 
df = pd.DataFrame({'Show': show,  'This Week Gross': weekblah, 'Diff $': diff, 'Avg Ticket': avg, 'Seats Sold': seats, 'Perfs': perfs, '% Cap': cap, 'Diff % cap': capdiff, 'Week': week,})
df[['Show','Theater']] = df['Show'].str.split('\n\n', expand=True)
#df[['This Week Gross', 'Potential Gross']] = df['This Week Gross'].str.split('\n', expand=True)
df[['Avg Ticket', 'Top Ticket']] = df['Avg Ticket'].str.split('\n', expand=True)
df[['Seats Sold', 'Seats in Theater']] = df['Seats Sold'].str.split('\n', expand=True)
df[['Perfs', 'Previews']] = df['Perfs'].str.split('\n', expand=True)   
```
